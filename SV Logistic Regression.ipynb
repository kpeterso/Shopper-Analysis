```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.tree import export_graphviz
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
df=pd.read_csv(r'D:\LocalData\peterk2\Desktop\Strat Vis Shopping Analysis\logistic regression data.csv',encoding = 'ISO-8859-1')
```

```python
df['Ethnic Group'].unique()
```

```python
clf1 = LogisticRegression()
clf2 = RandomForestClassifier()
clf3 = GaussianNB()

eclf = VotingClassifier(estimators=[('lr', clf1), ('rf', clf2), ('gnb', clf3)],voting='soft',weights=[1, 1, 1])
```

```python
dropVars = ['ID','SUV','New Model Purchased (Alpha Order)','New Model Purchased - Make/Model/Series (Alpha Order)','New Model Purchased - Brand','New Model - Corporation','New Model Segment','Age Bracket','Occupation','Education','Education Numeric','How Likely to Recommend Own Vehicle Brand Standardized','Purchase Price (100\'s)','Total Family Pre-Tax Income','State','Customer Region Classification','New Model Purchased - Region Origin','Segment Price Z score','Model Price Z score']
```

```python
dfData = df.drop(dropVars, axis=1)
#dfAAData = dfData[dfData['Ethnic Group']=='Hispanic/Latino']
```

```python
X = pd.get_dummies(dfData.drop('Sedan', axis=1))
y=dfData['Sedan']
```

```python
X
```

```python
X = X.apply(lambda x: x.fillna(x.mean()),axis=0)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)
```

```python
X_trainWeights = X_train['Sales Weighting']
X_train = X_train.drop('Sales Weighting', axis=1)
X_testWeights = X_test['Sales Weighting']
X_test = X_test.drop('Sales Weighting', axis=1)
```

```python
scores = [c.fit(X_train, y_train,X_trainWeights.values).score(X_test,y_test,X_testWeights.values) for c in (clf1, clf2, clf3, eclf)]
```

```python
b=pd.DataFrame(clf2.feature_importances_)
b['cols']=X_train.columns
b.sort_values(by=0,ascending=False)#.to_excel('Tree Importances.xlsx')
```

```python
importances = clf2.feature_importances_
std = np.std([tree.feature_importances_ for tree in clf2.estimators_],axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(X_test.shape[1]):
    print("%d. feature %s (%f)" % (f + 1, X_test.columns[indices[f]], importances[indices[f]]))

```

```python
# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X_test.shape[1]), importances[indices],color="r", yerr=std[indices], align="center")
plt.xticks(range(X_test.shape[1]), indices)
plt.xlim([-1, X_test.shape[1]])
plt.show()
```
